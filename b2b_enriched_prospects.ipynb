```python
import pandas as pd
import requests
import time
import ast

df = pd.read_json('dataset_y-combinator-scraper.json.json')

print("Loaded dataset with", len(df), "companies")
print("columns:", df.columns.tolist())

def is_b2b(row):
    tags = row.get('tags', [])
    if isinstance(tags, str):
        try:
            tags = ast.literal_eval(tags)
        except:
            tags = []
    if not isinstance(tags, list):
        tags = []
    tags_str = ' '.join(str(t).upper() for t in tags)

    desc = str(row.get('short_description', '')) + ' ' + str(row.get('long_description', ''))
    b2b_keywords = ['B2B', 'SAAS', 'SOFTWARE', 'AI', 'ENTERPRISE', 'DEVTOOLS', 'FINTECH', 'DEVELOPER', 'AUTOMATION', 'SECURITY', 'OPEN SSOURCE']
    return any(kw in tags_str for kw in b2b_keywords) or any(kw.lower() in desc.lower() for kw in b2b_keywords)
df_b2b = df[df.apply(is_b2b, axis = 1)].copy().reset_index(drop = True)
print(f"Qualified {len(df_b2b)} B2B/SaaS prospects")

df_b2b['priority_segment'] = pd.qcut(range(len(df_b2b)), q = 3, labels = [ 'High', 'Medium', 'Low'])

serper_key = "75c19a84743d6a2f190a3addc3d115529c299cde"

def enrich_company(row):
    name = row.get('company_name', 'Unknown')
    query = f"{name} official website instagram revenue employees founder linkedin"
    headers = {"X-API-KEY": serper_key, "Content_Type": "application/json"}
    payload = {"q": query, "num": 5}

    try:
        response = requests.post("https://google.serper.dev/search", headers=headers, json=payload, timeout=10)
        data = response.json()
        snippet = data.get("organic", [{}])[0].get("snippet", "No extra data")
        return snippet
    except Exception as e:
        return f"Enrichment failed: {str(e)[:80]}"
print("Enriching top 20 prospects...")
df_b2b['enriched_data'] = df_b2b.head(20).apply(enrich_company, axis = 1)

# save clay-ready files
df_b2b.to_json('prosect_list_for_clay.json', orient = 'records', indent = 2)
df_b2b.to_csv('yc_b2b_prospects_enriched.csv', index = False)

print("Done! Files saved in my notebook folder:")
print(" prospect_list_for_clay.json ready to import into Clay")
print(" yc_b2b_prospects_enriched.csv easy to open and view")

# Preview
print("\nPreview of first 5 qualified prospects:")
print(df_b2b[['company_name', 'priority_segment', 'enriched_data']].head())

            
```
